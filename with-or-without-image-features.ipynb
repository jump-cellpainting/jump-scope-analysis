```python
# How many image features are present in the feature select profile?

import glob
import os

def search_feature_selections(metadata_df, profile_path):
    """
    Reads <plate_name>.csv and <plate_name>_normalized_feature_select_negcon_batch.csv.gz
    and checks how many image features are present
    """
    df_list = []
    meta = metadata_df.copy()
    meta = meta[~meta["Batch"].str.contains("siteSub")]
    for ind, row in meta.iterrows():
        unselected_path = os.path.join(profile_path, row["Batch"], row["Assay_Plate_Barcode"], f"{row['Assay_Plate_Barcode']}.csv.gz")
        unselected_path = glob.glob(unselected_path)[0]
        unselected_df = pd.read_csv(unselected_path)
        unselected_img_feats = [i for i in unselected_df.columns if "Image" in i]

        selected_path = os.path.join(profile_path, row["Batch"], row["Assay_Plate_Barcode"], f"{row['Assay_Plate_Barcode']}_normalized_feature_select_negcon_batch.csv.gz")
        selected_path = glob.glob(selected_path)[0]
        selected_df = pd.read_csv(selected_path)
        selected_img_feats = [i for i in selected_df.columns if "Image" in i]

        print(f"--- {row['Assay_Plate_Barcode']} has {len(unselected_img_feats)} Image features and {len(selected_img_feats)} selected Image features ---")

        # for i in [x.strip() for x in row["channel_names"].split(",")]:
        #     num_feat_cols = [col for col in load_df.columns if i in col]
        #     meta.loc[ind, f"percent_{i}"] = (len(num_feat_cols) / len(load_df.columns)) * 100
    return metadata_df

experiment_df = pd.read_csv("output/all-profile-metadata.csv")
col_feats = search_feature_selections(experiment_df, "../jump-scope/profiles/")
```
