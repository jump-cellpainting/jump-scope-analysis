```python
import sys
sys.path.append("../../../") # Where to find utilssphering
import utilssphering
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read CSV
match_rep_df = pd.read_csv("../../../checkpoints/match_rep_df.csv")
match_rep_df
```

## Alternative figure legend

A - Assessment of magnification and aperture on ability to replicate profiles between wells treated with drugs with identical MoA (replicating) vs drugs with a similar MoA (matching)
B - Examining how number of sites taken per well impacts profile strength
C - All profiles were aggregated based on the number of sites used. Errorbar = SD
D - For profiles with 9 images per well, data was randomly subsampled to 4, 3, 2, and 1 site(s). Errorbar = SD
E - For the profiles examined in D, the cell count for each profile.


```python
# A

group = ["Magnification", "aperture"]

df_subset = match_rep_df[
                        (match_rep_df["sphering"] == True) &
                        (~match_rep_df["Batch"].str.contains("siteSub"))
                        ]

group_df = utilssphering.find_group_avg_df(df_subset, group)

fig, ax = plt.subplots(figsize=(5,5))

utilssphering.group_plot(
    df=group_df,
    x="Mean_Percent_Matching", 
    y="Mean_Percent_Replicating",
    group=group,
    error_x="SD_Percent_Matching",
    error_y="SD_Percent_Replicating",
    fig=fig,
    ax_=ax
)
utilssphering.group_plot(
    df_subset, 
    "Percent_Matching", 
    "Percent_Replicating", 
    group=group, 
    fig=fig, 
    ax_=ax, 
    legend=True)

# fig.savefig("(A-ALT)-rep-vs-match-mag_aperture.png")
```

```python
# B 
# Plot profile subsamples that originally had 9 images per well

group = "sites"

df_subset = match_rep_df[
                        (match_rep_df["sphering"] == True) &
                        (~match_rep_df["Batch"].str.contains("siteSub"))
                        ]

group_df = utilssphering.find_group_avg_df(df_subset, group)

fig, ax = plt.subplots(figsize=(5,5))

utilssphering.group_plot(
    df_subset, 
    "Percent_Matching", 
    "Percent_Replicating", 
    group=group, 
    fig=fig, 
    ax_=ax, 
    legend=True)

utilssphering.group_plot(
    df=group_df, 
    x="Mean_Percent_Matching", 
    y="Mean_Percent_Replicating",
    group=group,
    error_x="SD_Percent_Matching",
    error_y="SD_Percent_Replicating",
    fig=fig,
    ax_=ax
)
```
