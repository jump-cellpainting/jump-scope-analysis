```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
match_rep_df = pd.read_csv("../../../checkpoints/match_rep_df_with_evalzoo_proportions.csv")
match_rep_df = match_rep_df[~match_rep_df["Batch"].str.contains("siteSub")]
match_rep_df
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.regplot(match_rep_df["Percent_Replicating"], match_rep_df["ref_score"], ax=ax)
ax.set_xlabel("Percent Replicating", size=12)
ax.set_ylabel("Percent of MoA replicates above pvalue 0.05", size=12)

fig.savefig("(S2A)-replicating-vs-evalzoo-replicates.png", dpi=300, bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.regplot(match_rep_df["Percent_Matching"], match_rep_df["non_rep_score"], ax=ax)
ax.set_xlabel("Percent Matching", size=12)
ax.set_ylabel("Percent of MoA Matches above pvalue 0.05", size=12)

fig.savefig("(S2B)-matching-vs-evalzoo-moa_matches.png", dpi=300, bbox_inches="tight")
```
