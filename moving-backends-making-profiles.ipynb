```python
import os
import pandas as pd
import shutil
```

```python
metadata_df = pd.read_csv("/home/ubuntu/jump-scope/jump-scope-analysis/output/all-profile-metadata.csv")
metadata_df
```

```python


def move_profiles(metadata_df, source, destination):
    """Based on the provided metadata, retroactively move the backends to their respective locations prior
    to profile generation"""
    for root, dirs, files in os.walk(source):
        for ind, df in metadata_df.iterrows():
            for file in files:
                if file == df["Assay_Plate_Barcode"]+".csv.gz":
                    # print(
                    #     f"Moving: {os.path.join(root, file)} to: \n"
                    #     f'{os.path.join(destination, df["Batch"], df["Assay_Plate_Barcode"], file)}'
                    #     )
                    os.makedirs(os.path.join(destination, df["Batch"], df["Assay_Plate_Barcode"]), exist_ok=True)
                    # Move file to batch folder
                    src = os.path.join(root, file)
                    dst = os.path.join(destination, df["Batch"], df["Assay_Plate_Barcode"], file)
                    shutil.copy(src, dst)

move_profiles(metadata_df, 
            "/home/ubuntu/ebs_tmp/backend", 
            "/home/ubuntu/ebs_tmp/backend_reorganised")
```
