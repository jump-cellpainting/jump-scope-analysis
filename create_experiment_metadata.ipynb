```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
#col_order = ['Vendor','Batch','Plate_Map_Name','Assay_Plate_Barcode','Magnification','Objective_NA','Binning','Images_per_well','z_plane','Number_of_channels','Anomaly']
col_order = ['Vendor','Batch','Plate_Map_Name','Assay_Plate_Barcode','Binning','Number_of_channels','Anomaly']
        
```

```python
# Read barcode_platemap files
path = os.path.join('../jump-scope/metadata/platemaps/*/barcode_platemap.csv')
files = glob.glob(path)
df = pd.concat(
    (
        (
            pd.read_csv(_)
            .assign(Batch=_.split('/')[4])
            .assign(Vendor=_.split('/')[4].split('_')[1])
            .assign(Anomaly="none")
            .assign(Number_of_channels="5")
            .assign(Binning="1")
        ) for _ in files),
    ignore_index=True
)
df.head(50)
```

```python
for i, Vendor in enumerate(df.Vendor):
#    print(i,Vendor)
    if Vendor=='Yokogawa':
        df.loc[i,"Vendor"]=Vendor + '_' + df.loc[i, "Batch"].split('_')[2]
df.head(50) 
```

```python
for i, Vendor in enumerate(df.Vendor):
#    print(i,Vendor)
    if Vendor=='PE':
        Bin=df.loc[i, "Batch"].split('_')[2]
        if Bin=="Bin2":
            df.loc[i,"Binning"]="2"
            
df.head(50)
```

```python
# Sort rows and columns
df = df.sort_values(by=['Batch','Plate_Map_Name','Assay_Plate_Barcode'], ignore_index=True)[col_order]
```

```python
df.to_csv('output/experiment-metadata.tsv', index=False, sep='\t')
```
